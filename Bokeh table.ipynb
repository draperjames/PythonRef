```python
from os.path import dirname, join

import pandas as pd

from bokeh.io import push_notebook, show, output_notebook
from bokeh.layouts import row, widgetbox
from bokeh.models import ColumnDataSource, CustomJS
from bokeh.models.widgets import Slider, Button, DataTable, TableColumn, NumberFormatter
# from bokeh.io import curdoc
output_notebook()
```

```python
downloadjs="""var data = source.data;
var filetext = 'x,y\n';
for (i=0; i < data['x'].length; i++) {
    var currRow = [data['x'][i].toString(),
                   data['y'][i].toString().concat('\n')];

    var joined = currRow.join();
    filetext = filetext.concat(joined);
}

var filename = 'data_result.csv';
var blob = new Blob([filetext], { type: 'text/csv;charset=utf-8;' });

//addresses IE
if (navigator.msSaveBlob) {
    navigator.msSaveBlob(blob, filename);
}

else {
    var link = document.createElement("a");
    link = document.createElement('a')
    link.href = URL.createObjectURL(blob);
    link.download = filename
    link.target = "_blank";
    link.style.visibility = 'hidden';
    link.dispatchEvent(new MouseEvent('click'))
}
"""
```

```python
import numpy as np
import pandas as pd
```

```python
x = np.random.choice(list("ABCDEFG"),10)
y = np.arange(10)

s = pd.DataFrame([x,y]).T
s.columns = ["x","y"]

source = ColumnDataSource(s)


button = Button(label="Download", button_type="success")
button.callback = CustomJS(args=dict(source=source),code=downloadjs)


columns = [
    TableColumn(field="x", title="x"),
    TableColumn(field="y", title="y")
]

data_table = DataTable(source=source, columns=columns, width=800)

controls = widgetbox(button)
table = widgetbox(data_table)
show(row(controls,table))
```
