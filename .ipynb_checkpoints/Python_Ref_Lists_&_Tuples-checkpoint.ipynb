### Awesome quick way to replace elements in a list

```python
conditions = ["NonEx","Immediate Post","60 min Post","n/a"]
# dir(conditions)
super_names = ["Pool" if x == "n/a" else x for x in conditions]
super_names
```

```python
#Increment
listo = [1,2,3]
R = 0
for i in listo:
    R += i
R
```

```python
#Decrement
listo = [1,2,3]
R = 0
for i in listo:
    R -= i
```

```python
a = [1,2,3]
state=()  
for i in a:  
    state=state+(i,)
#Here is how to use increment correctly
state = ()  
for i in a:  
    state += (i,)
state
```
